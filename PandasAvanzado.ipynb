## Importando datos desde diferentes fuentes

```python
json = open('datos/alquiler.json')
print(json.read())
```

```python
#DataFrame de un archivo json
df_json = pd.read_json('datos/alquiler.json')
df_json
```

```python
txt = open('datos/alquiler.txt')
print(txt.read())
```

```python
#DataFrame de un archivo txt
df_txt = pd.read_table('datos/alquiler.txt')
df_txt
```

```python
#DataFrame de un archivo excel
df_xlsx = pd.read_excel('datos/alquiler.xlsx')
df_xlsx
```

```python
#DataFrame de un archivo html
df_html = pd.read_html('datos/datos_html_1.html')
df_html
```
