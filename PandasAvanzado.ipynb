# Filtrando datos

## Reporte de analisis de inmuebles residenciales

```python
datos = pd.read_csv('alquiler.csv', sep = ';')
datos.head(10)
```

```python
list(datos['Tipo'].drop_duplicates())
```

```python
residencial = ['Habitación',
 'Casa',
 'Departamento',
 'Casa en condominio',
 'Casa comercial',
 'Casa de villa']
residencial
```

```python
seleccion = datos['Tipo'].isin(residencial)
seleccion
```

```python
datos_residencial = datos[seleccion]
datos_residencial
```

```python
#Para verificar
list(datos_residencial['Tipo'].drop_duplicates())
```

```python
#Como los indices no estan secuenciales se hace lo siguiente
datos_residencial.index = range(len(datos_residencial))
datos_residencial
```

## Exportando base de datos

```python
datos_residencial.to_csv('alquiler_residencial.csv', sep = ';')
#Se exportó la columna index tambien
```

```python
datos_residencial_2 = pd.read_csv('alquiler_residencial.csv', sep = ';')
datos_residencial_2
#La columna 'Unnamed:0' es el index
```

# Organizando dataframes

```python
data = [[1,2,3],[4,5,6],[7,8,9]]
data
```

```python
list('321')
```

```python
df = pd.DataFrame(data, list('321'), list('ZYX'))
df
```

```python
#Ordenar el index
df.sort_index(inplace = True)
df
```

```python
#Ordenar las columnas
df.sort_index(inplace = True, axis = 1)
df
```

```python
#Ordenar los valores por la columna x
df.sort_values(by = 'X', inplace = True)
df
#Los indices se vuelven a desordenar
```

```python
#Ordenar los valores horizontalmente
df.sort_values(by = '3', axis = 1, inplace = True)
df
```
