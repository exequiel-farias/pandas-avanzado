# Selecciones y Frecuencias

```python
datos = pd.read_csv('alquiler_residencial.csv', sep = ';')
datos.head(10)
```

```python
#Seleccione solamente los inmuebles con tipo 'Departamento'
seleccion = datos['Tipo'] == 'Departamento'
n1 = datos[seleccion].shape[0] #Cant. o frecuencia de departamentos
n1
```

```python
#Seleccione los inmuebles clasificados con tipos 'Casa', 'Casa en condominio' y 'Casa de villa'
seleccion = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') |  (datos['Tipo'] == 'Casa de villa')
n2 = datos[seleccion].shape[0]
n2
```

```python
#Seleccione los inmuebles con area entre 60 y 100 metros cuadrados, incluyendo los limites
seleccion = (datos['Area'] >= 60) & (datos['Area'] <= 100)
n3 = datos[seleccion].shape[0]
n3
```

```python
#Seleccione los inmuebles que tengan por los menos 4 cuartos y alquiler menor a $2000.0
seleccion = (datos['Cuartos'] >= 4) & (datos['Valor'] < 2000.0)
n4 = datos[seleccion].shape[0]
n4
```

```python
print("N° de inmuebles clasificados con tipo 'Departamento' -> {}".format(n1))
print("N° de inmuebles clasificados con tipo 'Casa', 'Casa en codominio' y 'Casa de villa' -> {}".format(n2))
print("N° de inmuebles con area entre 60 y 100 metros cuadrados, incluyendo los limites -> {}".format(n3))
print("N° de inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2.000,0 -> {}".format(n4))
```

# Tratamiento de datos faltantes

```python
datos = pd.read_csv('alquiler_residencial.csv', sep = ';')
```

```python
datos.head(10)
```

```python
#Crea un dataframe donde si es true es porque tiene un valor null
datos.isnull()
```

```python
#Funcion parecida a la anterior pero trabaja de manera contraria
datos.notnull()
```

```python
#Para saber exactamente que columnas tienen valores nulos se hace lo siguiente
datos.info()
```

```python
#Como lo siguiente es una serie booleana, se puede convertir a condicion
datos['Valor'].isnull()
```

```python
#Me trae solamente los valores en lo que la condicion es true
datos[datos['Valor'].isnull()]
```

```python
A = datos.shape[0] #Cant de lineas antes de la eliminacion
datos.dropna(subset = ['Valor'], inplace = True) #Elimina filas en la que la columna 'Valor' tenga valores NaN
B = datos.shape[0] #Cant de lineas despues de la eliminacion
A - B #Cant de lineas eliminadas
```

```python
#Ahora no hay valores nulos
datos[datos['Valor'].isnull()]
```

## Tratamiento condicional

```python
#Podemos los valores nulos de la columna 'Mantenimiento' siempre y cuando el tipo sea 'Departamento'
datos[datos['Mantenimiento'].isnull()].shape[0] #Cant de lineas que tienen null en 'Mantenimiento'
```

```python
seleccion = (datos['Mantenimiento'].isnull()) & (datos['Tipo'] == 'Departamento' )
```

```python
A = datos.shape[0] #Cant de lineas antes de la eliminacion
datos = datos[~seleccion]
B = datos.shape[0] #Cant de lineas despues de la eliminacion
A - B #Cant de lineas eliminadas
```

```python
#Cant de lineas que todavia tienen null en 'Mantenimiento'
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
#Reemplazar los valores nulos restantes en 'Mantenimiento' e 'Impuesto' por 0
datos = datos.fillna({'Mantenimiento': 0, 'Impuesto': 0})
```

```python
#Cant de lineas que todavia tienen null en 'Mantenimiento'
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
#Cant de lineas que todavia tienen null en 'Impuesto'
datos[datos['Impuesto'].isnull()].shape[0]
```

```python
datos.info()
```

```python
#Exportar dataframe
datos.to_csv('alquiler_residencial.csv', sep = ';', index = False)
```
